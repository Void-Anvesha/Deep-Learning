```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post')
sequences
```

```python
model = Sequential()

#17->doc me kitni vocab hain/no.of unique words
#output_dim=2->hard word ko 2 number se represent karna chate h

model.add(Embedding(17,output_dim=2,input_length=5))

model.summary()
```
