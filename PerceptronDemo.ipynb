```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/placement-dataset (1).csv")
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(df['cgpa'],df['iq'],hue=df['placement'])
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
```

```python
p.fit(x,y)
```

```python
p.coef_ # Print the values of the weights
```

```python
p.intercept_ #Print the value of the bias
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values,y.values,clf=p,legend=2)
```
